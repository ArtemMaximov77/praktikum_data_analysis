**Описание проекта**  
В вашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений за несколько лет о продаже квартир в Санкт-Петербурге и соседних населённых пунктах.  
Ваша задача — выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, которые существуют на рынке недвижимости.  
О каждой квартире в базе содержится два типа данных: добавленные пользователем и картографические. Например, к первому типу относятся площадь квартиры, её этаж и количество балконов, ко второму — расстояния до центра города, аэропорта и ближайшего парка.

**Описание данных**  
`airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)  
`balcony` — число балконов  
`ceiling_height` — высота потолков (м)  
`cityCenters_nearest` — расстояние до центра города (м)  
`days_exposition` — сколько дней было размещено объявление (от публикации до снятия)  
`first_day_exposition` — дата публикации  
`floor` — этаж  
`floors_total` — всего этажей в доме  
`is_apartment` — апартаменты (булев тип)  
`kitchen_area` — площадь кухни в квадратных метрах (м²)  
`last_price` — цена на момент снятия с публикации  
`living_area` — жилая площадь в квадратных метрах (м²)  
`locality_name` — название населённого пункта  
`open_plan` — свободная планировка (булев тип)  
`parks_around3000` — число парков в радиусе 3 км  
`parks_nearest` — расстояние до ближайшего парка (м)  
`ponds_around3000` — число водоёмов в радиусе 3 км  
`ponds_nearest` — расстояние до ближайшего водоёма (м)  
`rooms` — число комнат  
`studio` — квартира-студия (булев тип)  
`total_area` — общая площадь квартиры в квадратных метрах (м²)  
`total_images` — число фотографий квартиры в объявлении

### Найдите и изучите пропущенные значения в столбцах:

Если логичную замену предложить невозможно, то оставьте эти значения пустыми. Пропуски — тоже важный сигнал, который нужно учитывать.

```python
(100*df[df.columns].isnull().sum()/len(df)).sort_values(ascending=False).round(2)
```

Имеются небольшие пропуски в названии населенного пункта `locality_name` и этажность дома `floors_total` .   
Наибольшее количество пропусков в колонке аппартаменты `is_apartment`. В данном случае все пропуски можно заменить на значение `False`, т.к. аппартаменты это нежилая комерческая недвижимость и таких продаж должно быть мало.  
Много значений отсуствует в локационных данных таких как число парков, водоемов по близости, расстояние до центра, аэропорта, ближайшего парка, водоема. Эти данные оставим как есть.  
Оставшиеся пропуски попробуем заменить.  
Причины пропусков могут быть следующие:
 - необязательность заполнения отдельных полей при регистрации объявления; 
 - потеря данных при выгрузке; 
 - сокрытие некоторой информации...

Так, например, группа локационных данных возможно подсчитывается автоматически, при условии локации объекта недвижимости только в Питере. Отсутствие данных о количестве балконов, возможно связано с их физическим отсутствием. Отсутствие площади кухни, возможно при условии, если квартира студия. Про аппартаменты мало кто вообще знает, что это такое. Высоту потолков возможно специально не заполняют, чтобы скрыть низкие значения...

```python
# Там где количество балконов не указано меняем значения на 0
df['balcony'] = df['balcony'].fillna(0)
```

```python
# Если не указана этажность дома, то возможно указанный этаж является последним
df['floors_total'] = df['floors_total'].fillna(df['floor'])
```

```python
# Отсутствующие значения в колонке аппартаменты заменим на False
df['is_apartment'] = df['is_apartment'].fillna(False)
```

```python
# посмотрим на локационные данные, возможно они отсутствуют в случае, если квартира находится за пределами Питера.
df[['locality_name', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 
    'ponds_around3000', 'ponds_nearest']].query('locality_name != "Санкт-Петербург"').isnull().sum()

# Так и есть!
```

```python
# С высотой полков пока непонятно, посмотрим на уникальные значения
df['ceiling_height'].unique()
```

```python
# статистика
df['ceiling_height'].plot(kind='box', ylim=(1.5, 3.5));
df['ceiling_height'].describe()
```

```python
# Среднее и медианное значения находятся рядом. Принимаем решение, заменить отсутствующие значения на 2.7 м.
df['ceiling_height'] = df['ceiling_height'].fillna(2.7)
```

```python
# Остались living_area, kitchen_area, days_exposition.... 
# Думаю оставить их как есть, т.к. процент отсутствующих значений незначительный

(100*df[['living_area', 'kitchen_area', 'days_exposition']].isnull().sum()/len(df)).sort_values().round(0)
```

Пропуски в данных были заменены в следующих колонках:
- `balcony`, на значение 0;
- `floors_total`, на значение номера этажа;
- `is_apartment`, на значение False;
- `ceiling_height`, на значение 2.7.
